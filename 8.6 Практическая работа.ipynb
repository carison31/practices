**Способ №3 (выполните по желанию)**. Это самое эффективное и `элегантное` решение. Существует такая логическая операция `xor` или исключающее "или". 

"Логическая" значит, что она выполняется над битами числа в двоичной системе исчисления. Берётся первый бит (это 1 или 0), затем второй бит, и с этими двумя битами выполняется  логическая операция, результатом которой будет определенный третий бит. Каждый бит принимает только два значения, 1 или 0, поэтому для каждой логической операции возможны только четыре комбинации операндов и результата, которые в совокупности называются `таблицей истинности` этой логической операции. У логической операции `xor` таблица истинности такая:

```bash
0 xor 0 = 0
1 xor 0 = 1
0 xor 1 = 1
1 xor 1 = 0
```

Результат `xor` равен единице, только если ее операнды – разные. Если операнды одинаковые, то результат `xor` равен нулю. 

У этого свойства есть интересное следствие. Возьмём два одинаковых числа, например, `5` и `5`, и запишем их в двоичной системе исчисления. Запишем в столбик, одно число под другим.

```bash
101 
101
```

Числа одинаковые, поэтому в двоичной форме они тоже выглядят одинаково. Теперь последовательно применим `xor` к битам наших чисел: первый бит _поксорим_ с первым, второй – со вторым, и так далее. Так как числа одинаковые, биты тоже будут одинаковыми, поэтому результат `xor` будет равен нулю. Биты `одинаковых` чисел зануляют друг друга. Получается, что если одно число `поксорить` само на себя, то будет `0`: `5 xor 5 = 0`, `10 xor 10 = 0`, `42 xor 42 = 0`

Теперь дальше, давайте `поксорим` `0` с каким-нибудь числом, например, с "5". Для этого запишем числа в столбик в двоичной форме:

```bash
000
101
```

Все биты нуля ожидаемо нулевые, у пятерки на некоторых позициях есть единицы. `xor` устроен так, что для разных битов он даёт `1`, а для одинаковых – `0`. У нуля и любого другого числа одинаковые биты могут быть только нулевыми (`0 xor 0 = 0`), а разные – только если у другого числа на соответсвующих позициях стоят единицы (`0 xor 1 = 1`). Получается, что нулевые биты другого числа нуль оставляет нулями, а единичные – единицами. Другими словами, ноль, `поксоренный` с любым другим числом, не изменяет это другое число: `0 xor 5 = 5`, `0 xor 10 = 10`, `0 xor 42 = 42`.

В Python есть специальный оператор, который выполняет `xor` операндов – это `^`. 

```bash
42 ^ 42 = 0
0  ^ 42 = 42
42 ^ 0  = 42
```

Теперь вы знаете всё, чтобы эффективно решить задачу с поиском одинокого числа очень, очень эффективно. Это решение "со звездочкой", поэтому здорово, если у вас получится!

````python
# Первый способ медленный

list2 = [0, 1, 3, 4, 7, 98, 3, 1, 98, 7, 16, 16, 4, 42, 0]

for i in range (len(list2)):
    counter = 0
    for j in range (len(list2)):
        if list2[i] == list2[j]:
            counter += 1
        elif counter == 2:
            break
    if counter == 1:
        print (list2[i])
            
# Второй способ побыстрее (почти как freqs способ)
check_list = [] # список, в который мы будем учитывать уже встретившиеся раннее числа 

for i in range (len(list2)):
    for j in range (i + 1, len(list2)):
        if list2[i] == list2[j]:
            check_list.append(list2[i])
            break
        elif list2[i] in check_list:
            break
    if list2[i] not in check_list:
        print (list2[i])
````

````python
list4 = [2, 12, 7, 8, 0]

result = []
window_size = 3
# Ваш код здесь

for i in range (len(list4) - window_size + 1):
    summary = 0
    for j in range (0, window_size):
        summary += list4[i + j]
    average = summary / window_size
    result.append(average)
print (result)
    
````
