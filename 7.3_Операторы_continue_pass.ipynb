## Операторы continue, pass

Напишем цикл, который для каждого дня рабочего месяца будет выводить его номер. Чтобы вывод красивее выглядел, будем использовать конкатенацию, а для этого создадим новую переменную, которая будет равна номеру дня, но при этом будет строкой.


```python
for day in range(1, 32):
    day_str = str(day)
    print("Сегодня " + day_str + "-ое число")
```

Мы говорим о рабочем месяце, поэтому добавим результат каждого рабочего дня – усталость. В строчку с числом добавим параметр `end`, равный строке с тире. Этот параметр задаёт, чем заканчивается вывод. По умолчанию это перевод строки, но в данном случае нам будет удобнее разделять две фразы с помощью тире.


```python
for day in range(1, 32):
    day_str = str(day)
    print("Сегодня " + day_str + "-ое число", end = " - ")
    print("Что-то я устала")
```

Получилось достаточно депрессивно, потому что мы забыли про выходные.

Каждое воскресенье мы спим весь день и поэтому не устаём. Нужно добавить условие, при котором каждый седьмой день не приводил бы к усталости. 

Само условие мы зададим с помощью остатка от деления. Для простоты будем считать, что этот месяц начинался с понедельника. Если воскресенье – седьмой день и все последующие воскресенья идут через семь дней, то остаток при делении на 7 будет равен нулю. 

А дальше мы не хотим выводить фразу про усталость, поэтому будем использовать команду `continue`. Она сообщает, что надо пропустить абсолютно всё, что написано в цикле после этого момента, и сразу же перейти к следующему повторению цикла, то есть в его начало.

Во все остальные дни мы усердно работаем, добавим это в раздел `else`


```python
for day in range(1, 32):
    day_str = str(day)
    print("Сегодня " + day_str + "-ое число", end = " - ")
    if day % 7 == 0:
        print("сплю весь день")
        continue
    else:
        print("усердно работаю", end = " - ")
    print("Что-то я устала")
```

Первые 6 чисел шли по циклу, проверяли условие деления (оно не выполнялось), поэтому  переходили к разделу `else`, печатали фразу про работу, а дальше переходили к  фразе  про усталость, стоящей после блока с условием. 

Седьмое число выполнило условие про деление, сообщило про сон, а дальше пропустило фразу про усталость и перешло сразу к восьмому числу. То же самое произошло с 14-м числом.

Добавим еще одно условие. По субботам мы занимаемся домашним хозяйством, это скучно, что даже сообщить нечего, а к концу дня снова устаём. Давайте будем проскакивать этот день сразу к усталости. Для этого у нас есть команда `pass`. Она сообщает пропустить блок, не делать ничего. А раз ничего делать не надо, то следующий шаг будет сделан как обычно.


```python
for day in range(1, 32):
    day_str = str(day)
    print("Сегодня " + day_str + "-ое число", end = " - ")
    if day % 7 == 0:
        print("сплю весь день")
        continue
    elif day % 7 == 6:
        pass
    else:
        print("усердно работаю", end = " - ")
    print("Что-то я устала")
```

Теперь мы можем видеть разницу трёх условий.

1. С первого по пятое число сначала выполняется первая строка. Потом проверяется условие, оно попадает в раздел `else` и выполняет инструкцию из него. Потом выполняется та часть цикла, которая стоит после условий – сообщение об усталости. После этого идет на новый виток цикла.
2. Когда наступает шестое число, выполняется первая строка. Потом проверяется условие, и оно попадает под второй вариант – `elif`. Он сообщает, что нужно просто пропустить и пойти дальше, поэтому переходим к инструкциям после условия – пишем про усталость. И на новый виток.

3. Когда наступает седьмое число, как обычно выводится первая строка. Потом проверяется условие, идет по первому пути `if`. Печатает сообщение про сон, а потом, следуя команде `continue`, пропускает всё, что в цикле дальше (включая сообщение про усталость) и просто запускает следующий виток цикла.

Команду `pass` удобно использовать в процессе написания программ, когда вам нужно, чтобы программа запускалась, но вы ещё не прописали все действия. Оставить блок совсем пустым вы не можете, программа выдаст ошибку – а с командой `pass` запустить получится.

Давайте рассмотрим ещё один пример, но в этот раз будем использовать переменную из диапазона цикла не только как условие, но и в расчётах. В этот раз подойдём схематично. Мы хотим пройтись по всем числам до 10, при этом все нечётные пропустить, а для всех чётных вывести его квадрат. В конце цикла добавим фразу, чтобы отслеживать ход программы.



```python
for i in range(1, 11):
    if i % 2 == 1:
        continue
    else:
        print(i ** 2)
    print(i, " - четное число")
```

В этом примере более наглядно видно, что в случае нечётных чисел программа идёт по первому пути, встречается с `continue` и сразу идёт в начало цикла. В случае чётных – идёт по второму пути, печатает квадрат числа, а потом печатает и то, что после условия, и только потом возвращается в начало цикла.
